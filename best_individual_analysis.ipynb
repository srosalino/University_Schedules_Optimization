# Create dataframe where each Row is a Turn and each Column is a Block - from Monday_Block1 to Friday_Block8

```python
# Combine the blocks for each day into a single row
combined_data = [sum(day, []) for day in best_individual]

# Define the column names
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
columns = [f"{day}_Block{i+1}" for day in days for i in range(8)]

# Create the DataFrame
df = pd.DataFrame(combined_data, columns=columns)
```

# Count of Blocks per Subject for each Turn, and Breaks

```python
# List of unique subjects including 'Break'
unique_subjects = set()
for turn in combined_data:
    unique_subjects.update(turn)

# Count the occurrences of each subject in each Turn
subject_counts = []
for turn in combined_data:
    turn_count = {subject: turn.count(subject) for subject in unique_subjects}
    subject_counts.append(turn_count)

# Display the counts in a readable format
subject_counts_df = pd.DataFrame(subject_counts)
subject_counts_df = subject_counts_df.fillna(0).astype(int)
```

# Days without Breaks for each Turn

```python
# Define a function to count the number of days without 'Break' for each turn
def count_days_without_break(turn):
    days_without_break = 0
    for i in range(0, 40, 8):  # Iterate through each day (8 blocks)
        if 'Break' not in turn[i:i+8]:
            days_without_break += 1
    return days_without_break

# Apply the function to each turn and store the results
days_without_break_counts = [count_days_without_break(turn) for turn in combined_data]

# Display the results
days_without_break_counts
```

# Overlaps

```python
overlaps = {}
for col in df.columns:
    value_counts = df[col].value_counts()
    for value, count in value_counts.items():
        if value != "Break" and count > 1:
            if col not in overlaps:
                overlaps[col] = []
            overlaps[col].append(value)

# Display the overlaps
overlaps
```
